```python
import matplotlib.pyplot as plt
data=pd.read_csv('/kaggle/input/breast-cancer-wisconsin-data/data.csv')
```

```python
data.drop(['id','Unnamed: 32'],axis=1 , inplace=True)
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
def binary_conversion(r):
    if r=='M':
        r=1
    else:
        r=0
    return r    
```

```python
data['diagnosis']=data['diagnosis'].apply(binary_conversion)
```

```python
data
```

```python
import seaborn as sns
```

```python
sns.lmplot(x='radius_mean' ,y='texture_mean' , data=data , hue='diagnosis')
```

```python
x=data.iloc[:,1:]
label=data.iloc[:,0]
```

```python
x
```

```python
label
```

```python
x[x=='None'].count()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain , ytest = train_test_split(x , label , test_size=0.2 , random_state=1 , shuffle=label)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(xtrain , ytrain)
```

```python
knn.score(xtrain,ytrain)
```

```python
predicted=knn.predict(xtest)
```

```python
knn.score(xtest,predicted)
```

```python
knn.score(xtest,ytest)
```
